```python
def binary_search(list, x):
    low = 0
    high = len(list) - 1
    mid = 0
    while low <= high:
        mid = (high + low) // 2
        if list[mid] < x:
            low = mid + 1
        elif list[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1
list=[45,55,67,76,88 ]
x=67
print(binary_search(list,x))
```

```python
#7.Write a python program to implement the various operations for Stack ADT i.) Push .) Pop iii.) Display.
#with class
class Stack:
    def __init__(self):
        self.items=[]

    def push(self,item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def is_empty(self):
        return len(self.items)==0

    def display(self):
        if not self.is_empty():
            print("Stack contents:")
            for item in reversed(self.items):
                print(item)
        else:
            print("Stack is empty")

stack=Stack()
while True:
    print("1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Quit")
    choice=input("Enter your choice (1/2/3/4):")
    if choice=='1':
        item=input("Enter the item to push:")
        stack.push(item)
        print(f"{item} pushed to the stack.")
    elif choice=='2':
        popped_item=stack.pop()
        print(f"Popped item: {popped_item}")
    elif choice=='3':
        stack.display()
    elif choice=='4':
        print("End statment")
        break
    else:
        print("Invalid choice")
```

```python
#8. Write a python script to implement the various operations for Queue ADT1.) Insert ii.) Delete iii.) Display.
#with class
class Queue:
    def __init__(self):
        self.items=[]

    def insert(self,item):
        self.items.append(item)

    def delete(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return "Queue is empty"

    def is_empty(self):
        return len(self.items)==0

    def display(self):
        if not self.is_empty():
            print("Queue contents:")
            for item in self.items:
                print(item)
        else:
            print("Queue is empty")

queue=Queue()

while True:
    print("1. Insert")
    print("2. Delete")
    print("3. Display")
    print("4. Quit")

    choice=input("Enter your choice:")
    if choice=='1':
        item=input("Enter the item to insert: ")
        queue.insert(item)
        print(f"{item} inserted into the queue")
    elif choice=='2':
        deleted_item=queue.delete()
        print(f"Deleted item: {deleted_item}")
    elif choice=='3':
        queue.display()
    elif choice=='4':
        print("End Statement")
        break
    else:
        print("Invalid choice")
```

```python
#9.infix to postfix a)A/B^C+D*E-F*G
class Stack:
    def __init__(self):
        self.items=[]

    def push(self,item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items)==0

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def infix_to_postfix(infix):
    precedence={'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    postfix=[]
    stack=Stack()

    for char in infix:
        if char.isalnum():
            postfix.append(char)
        elif char=='(':
            stack.push(char)
        elif char==')':
            while not stack.is_empty() and stack.peek() != '(':
                postfix.append(stack.pop())
            stack.pop()  # Pop '(' from the stack
        else:
            while not stack.is_empty() and precedence.get(char, 0) <= precedence.get(stack.peek(), 0):
                postfix.append(stack.pop())
            stack.push(char)

    while not stack.is_empty():
        postfix.append(stack.pop())

    return ''.join(postfix)

infix="A/B^C+D*E-F*G"
postfix=infix_to_postfix(infix)
print("Infix:", infix)
print("Postfix:", postfix)

```

```python
#6.singly linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def find_smallest(self):
        if not self.head:
            return None

        current = self.head
        smallest = current.data

        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next

        return smallest

    def insert_if_not_duplicate(self, data):
        if not self.head:
            self.head = Node(data)
            return

        current = self.head

        while current:
            if current.data == data:
                return  # Data already exists, don't insert
            if current.next is None:
                current.next = Node(data)
                return
            current = current.next

    def display_reverse(self, node):
        if node is None:
            return
        self.display_reverse(node.next)
        print(node.data, end=' ')

# Create a linked list
linked_list = LinkedList()

# Create a list (a)
elements = [10, 5, 15, 20, 10, 5]
for element in elements:
    linked_list.insert(element)

# Find the smallest element (b)
smallest = linked_list.find_smallest()
if smallest is not None:
    print("Smallest element:", smallest)
else:
    print("List is empty")

# Insert an element if it's not a duplicate (c)
new_element = int(input("Enter an element to insert if not a duplicate: "))
linked_list.insert_if_not_duplicate(new_element)

# Display the elements in reverse order (d)
print("Elements in reverse order:")
linked_list.display_reverse(linked_list.head)
```
