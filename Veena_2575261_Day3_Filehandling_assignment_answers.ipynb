```python
#7v)exception handling
def word(inputfile, outputfile):
    try:
        with open("C:/Users/Veena/Desktop/input.txt",'r') as inputfile:
            inputfile.read()
        with open("C:/Users/Veena/Desktop/files/output.txt", 'w') as outputfile:
            print(f'Successfully saved to{outputfile}')

    except FileNotFoundError:
        print(f'Error: File "{inputfile}" not found.')
    except Exception as e:
        print(f'Error: An exception occurred - {str(e)}')

# Example usage:
try:
    word('input.txt', 'output.txt')
except KeyboardInterrupt:
    print('Operation aborted by user.')
```

```python
#4log file
from collections import Counter

# Function to analyze a log file
def analyze_log(log_file):
    try:
        with open("C:/Users/Veena/Desktop/files/logfile.txt",'r') as file:
            lines = file.readlines()

        # Extract specific information
        user_ids = []
        actions = []

        for line in lines:
            parts = line.strip().split('\t')
            print(parts)
            if len(parts) >= 3:
                timestamp, user_id, action = parts[:3]
                user_ids.append(user_id)
                actions.append(action)

        # Calculate the number of unique users
        unique_users = len(set(user_ids))
        print(actions)

        # Find the most common action
        most_common_action = Counter(actions).most_common(1)[0][0]

        print(f"Number of unique users: {unique_users}")
        print(f"Most common action: {most_common_action}")

    except FileNotFoundError:
        print(f"Error: '{log_file}' not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Example usage:
log_file ='C:/Users/Veena/Desktop/files/logfile.txt'
analyze_log(log_file)
```

```python
#3create function sed ,two files input and output,copy contents to outputfile ,replace string
def sed(pattern, replacement, inputfile, outputfile):
    try:
        with open("C:/Users/Veena/Desktop/files/inputfile.txt", 'r') as inputfile:
            content = inputfile.read()
        modified_content = content.replace(pattern, replacement)
        with open("C:/Users/Veena/Desktop/files/outputfile.txt", 'w') as outputfile:
            outputfile.write(modified_content)
        
        print(f'Successfully replaced pattern and saved to {outputfile}')
    except FileNotFoundError:
        print("File not found")
    except Exception as e:
        print(f'Error: An exception occurred - {str(e)}')
try:
    sed('oldpattern', 'newpattern', 'inputfile.txt', 'outputfile.txt')
except KeyboardInterrupt:
    print('Operation aborted by user.')
```
